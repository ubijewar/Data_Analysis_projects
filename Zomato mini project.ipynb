```python
dp.drop(['Country Code','Locality','Longitude','Latitude',],axis=1,inplace=True)
dp.head(5)
```

```python
dp1=dp.loc[(dp.City=='Indian Rupees(Rs.)')]
dp1.head()
```

```python
dp1=dp.loc[(dp.City=='New Delhi')]
dp1.tail(4)
```

```python
dp1.groupby('Restaurant Name')['Votes'].max().nlargest(10).plot.bar()
plt.title('High Rating Restaurant');
```

```python
dp1['Rating text'].value_counts().plot(kind='pie',autopct='%1.1f%%',explode=[0.1,0.1,0.1,0.1,0.6,0.4])
plt.title('Rating Text Of Restaurant of (New Delhi)');
```

```python
dp['Rating text'].value_counts().plot(kind='pie',autopct='%1.2F%%',explode=[0.1,0.1,0.1,0.1,0.2,0.4])
plt.title('Rating text of Restaurant of (overall)');
```
